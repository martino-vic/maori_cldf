```python
#create forms.csv fromr mri3.csv
import pandas as pd

def forms():
    dfm = pd.read_csv("mri3.csv")
    #return dfm
    dfforms = pd.DataFrame()
    for form, donor in zip(["mri", "L1_ipa"], ["en", "L2_ipa"]):
        dfforms[form] = list(dfm[form]) + list([i for i in dfm[donor] if isinstance(i, str)])

   # for meaning in ["mri"]:
    #    dfforms[meaning] = list(dfm[meaning]) + list(["" for i in dfm[donor] if isinstance(i, str)])

    dfforms["Language_ID"] = ["0"]*len(list(dfm[donor])) + list(["1" for i in dfm[donor] if isinstance(i, str)])
    
    dfforms.reset_index(inplace=True)
    dfforms = dfforms.rename(columns={"index":"ID"})
    
    dfforms = dfforms.rename(columns={"mri":"Form", "L1_ipa":"IPA"})
    dfforms.to_csv("forms.csv", encoding="utf-8", index=False)

def borrowings():
    dfm = pd.read_csv("mri3.csv")
    dfm.reset_index(inplace=True)
    dfm = dfm.rename(columns={"index":"ID"})

    dfborr = pd.DataFrame()
    dfborr["Target_Form_ID"] = [i for i,j in zip(dfm["ID"],dfm["en"]) if isinstance(j, str)]
    dfborr["Source_Form_ID"] = [i for i in range(len(dfm),len(dfm)+len(dfborr))]

    dfborr.reset_index(inplace=True)
    dfborr = dfborr.rename(columns={"index":"ID"})
    dfborr.to_csv("borrowings.csv", encoding="utf-8", index=False)

forms()
borrowings()
```
